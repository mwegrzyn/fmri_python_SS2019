### modul zum Anzeigen von MRT-Bildern laden

```python
from nilearn import plotting
```

Wir k√∂nnen das Bild jetzt anzeigen

```python
plotting.plot_anat(mpr_file)
```

Wir k√∂nnen definieren, welche Koordinaten wir zeigen wollen

Wir k√∂nnen mehrere Schichten der gleichen Ansicht anzeigen

```python
plotting.plot_anat(mpr_file, # meine datei
                   display_mode='x', # ansicht
                   cut_coords=[-40,-20,20,40] # schnitte
                  )
```

```python
plotting.plot_anat(mpr_file, # meine datei
                   display_mode='y', # ansicht
                   cut_coords=[-40,-20,20,40] # schnitte
                  )
```

### das Ganze mit einer for-Schleife automatisieren

Beispiel:

```python
meine_zahlen = [1,2,3] # Liste mit 3 Eintr√§gen
```

```python
meine_zahlen
```

```python
type(meine_zahlen)
```

```python
for diese_zahl in meine_zahlen: # f√ºr jeden Eintrag der Liste
    print(diese_zahl) # Zeige den Eintrag
```

Liste mit Ansichten:

```python
meine_ansichten = ['x','y','z']
```

```python
meine_ansichten
```

Laden von Matplotlib, damit wir nach jedem Durchgang der Schleife einen neuen Plot angezeigt kriegen

```python
import matplotlib.pyplot as plt
```

display_mode wir jetzt durch den jeweiligen Wert von der Schleifen-Variable "diese_ansicht" definiert:

```python
for diese_ansicht in meine_ansichten:
    print(diese_ansicht)
    plotting.plot_anat(mpr_file, # meine datei
                   display_mode=diese_ansicht, # ansicht
                   cut_coords=(-40,-20,20,40) # schnitte
                  )
    plt.show()
```
